# MatplotLib

- Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
- It provides an object oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major pros of Matplotlib are:
- generally easy to get started for simple plots
- support for custom labels and texts
- Great control of every element in a figure
- high-quality output in many formats
- very customisable in general

### Seaborn is used more for plotting because it aids in statistical analysis .

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# If you use '%matplotlib inline', you won't have to use plt.show() everytime. It'll automatically show the plot.
# ONLY applicable in Jupyter
```

```python
import numpy as np
```

```python
# Examples
# Since we need 2-D graphs, we'll need atleast 2 variables

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# Plotting using matplotlib

#plt.scatter() -> To plot a scatter plot of the values provided

plt.scatter(x,y,c='r') # c is for scatter points' color ('r','g','b')
```

```python
# How to name the X and Y axes ?

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2-D')
```

```python
# Saving the plot image

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2-D')
plt.savefig('Plot-test.png')
```

```python
# plt.plot() -> Plots points as a line

y = x*x
plt.plot(x,y)
```

```python
plt.plot(x,y,'r') # 3rd parameter is for color
```

```python
plt.plot(x,y,'r--') # r-- for dashed line
# r*- for line with points marked
# ro for scatter
# ro-- for scatter with dashed line
# can also do the same using other parameters
```

```python
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2,markersize=12)
```

```python
# Creating subplots -> Multiple plots within one diagram
# plt.subplot(rows,columns,position)

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
```

```python
x = np.arange(1,11)
y = 2*x+5
plt.plot(x,y)
```

```python
# accessing value of pi
np.pi
```

```python
# compute x and y coordinates for points on sine curve
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('Sine waveform')
plt.plot(x,y)
```

```python
# subplot()

# compute x and y coordinate points on sine and cosine curves
x = np.arange(0,5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# set up a subplot grid with height 2 and width 1
plt.subplot(2,1,1) # first plot
plt.plot(x,y_sin,'r--')
plt.title('Sin')

# Second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cos')
```

## Histograms

```python
from numpy.random import randint
a = randint(1,90,15)
print('a =',a)
plt.hist(a) # default bins = 10
# plots density of nos. in a range
```

## Boxplot

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
# std is for standard deviation -> 1st b/w 0 to 1, then 0 to 2, then 0 to 3
data
```

```python
plt.boxplot(data, vert = True, patch_artist = True) 
# If patch_artist is False, the box won't be colored
# If vert is False, then plot will be in horizontal direction
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210] # Cumulative value -> so that calculated based on %ile
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0.1)  # explode 1st and 4th slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
```
